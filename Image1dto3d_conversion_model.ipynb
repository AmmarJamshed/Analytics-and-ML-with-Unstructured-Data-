```python
import numpy as np
import cv2
```

In Taking Inspiration of my Brother In laws project I decided to make a little project of my own for fun - A ML model which can convert 1 dimensional images to 3 dimesnional images.

## Function to Convert images

```python
def convert_1d_to_3d_image(image_1d, width, height):
    # Reshape 1D image array to 2D image array
    image_2d = np.reshape(image_1d, (height, width))
    
    # Convert 2D image to 3D image
    image_3d = cv2.cvtColor(image_2d, cv2.COLOR_GRAY2BGR)
    
    return image_3d
```

## Random Image data test

```python
# Example usage
width = 100
height = 100
```

```python
# Generate a random 1D image array
image_1d = np.random.randint(2, 256, width * height)
```

```python
# Convert 1D image to 3D image
image_3d = convert_1d_to_3d_image(image_1d, width, height)
```

```python
# Flatten the image into a 1D array
image_1d = image.flatten()

# Convert 1D image to 3D image
image_3d = convert_1d_to_3d_image(image_1d, width, height)
```
